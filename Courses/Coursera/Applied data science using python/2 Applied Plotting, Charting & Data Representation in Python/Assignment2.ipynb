```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))
    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(100,'4e86d2106d0566c6ad9843d882e72791333b08be3d647dcae4f4b110')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import matplotlib.dates as dates
def data_prep():
    df = pd.read_csv('data/C2A2_data/BinnedCsvs_d100/4e86d2106d0566c6ad9843d882e72791333b08be3d647dcae4f4b110.csv')
    df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
    df['Year'] = df.Date.dt.year

    df['DayOfYear'] = df['Date'].dt.strftime('%m/%d') #df.Date.dt.strftime('%b') + '-' + df.Date.dt.strftime('%d') 

    # Filter out leap days (i.e. February 29th)
    df = df.loc[(~((df.Date.dt.is_leap_year) & (df.Date.dt.month ==2) & (df.Date.dt.day ==29)))]

    # Readd all dates.  All will have year 1900, and this doesn't matter since year was extracted
    df['DayOfYear'] = pd.to_datetime(df['DayOfYear'], format='%m/%d')
    
    # Sort by the Date column
    df = df.sort_values('Date')

    df['Temperature'] = df['Data_Value']/10

    # Drop the Data_Value and Date columns
    for col in ['Data_Value']: #'Date', 
        df = df.drop(col, axis=1)
    return df
```

```python
def get_min_max_pre_2015():
    df = data_prep()
    df = df[df.Year < 2015]
    
    # Get the max and min based on DayOfYear
    result = df.groupby('DayOfYear')['Temperature'].agg(['max', 'min']).reset_index()
    
    # Create float representation of DayOfYear
    result['fds'] = result['DayOfYear'].apply(lambda x: dates.date2num(x))
    
    return result
```

```python
def get_min_max_2015():
    
    """
    Get the record highs and lows that occurred in 2015
    """
    
    pre2015 = get_min_max_pre_2015()
    df = data_prep()
    df = df[df.Year == 2015]
        
    # Get the max and min based on DayOfYear in 2015
    df2015 = df.groupby('DayOfYear')['Temperature'].agg(['max', 'min']).rename(columns={'max': 'max2015', 'min': 'min2015'}).reset_index()
    
    # Merge the pre2015 and 2015 DataFrames
    result = pd.merge(pre2015, df2015, how='outer', left_on='DayOfYear', right_on='DayOfYear')
    
    # Determine if min or max is more extreme than pre2015
    result['min_record_in_2015'] = result['min2015'] < result['min']
    result['max_record_in_2015'] = result['max2015'] > result['max']
    
    # Remove the rows with both values False
    result = result.loc[(result.min_record_in_2015 == True) | (result.max_record_in_2015 == True)]
    
    return result
```

```python
%matplotlib notebook

# Create new figure
plt.figure()

# Plot the record low and high from 2005-2014
pre2015 = get_min_max_pre_2015()
plt.plot(pre2015['fds'], pre2015['min'], color='red', lw=.5, label='Min for the day between 2005-2014')
plt.plot(pre2015['fds'], pre2015['max'], color='blue', lw=.5, label='Max for the day between 2005-2014')

# Plot the scatter points of records broken in 2015
df2015 = get_min_max_2015()


plt.scatter(df2015['fds'], df2015['min'], color='blue', marker='o', s=10, label='Min in 2015')
plt.scatter(df2015['fds'], df2015['max'], color='black', marker='o', s=10, label='Max in 2015')

ax = plt.gca()

# Set date axis format
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))
ax.xaxis.set_major_locator(dates.MonthLocator())
loc, labels = plt.xticks()
plt.setp(labels, rotation=90, fontsize=7)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.legend(loc='lower center', fontsize=7)
plt.xlabel('Day of the Year',fontsize=10)
plt.ylabel('Temperature in Celcius (Degree)', fontsize=10)
plt.title('Record temperatures for each day of the year', fontsize=11)

# fill the area between the record low and highs from 2005-2014
plt.gca().fill_between(pre2015['fds'], 
                       pre2015['min'], pre2015['max'], 
                       facecolor='blue', 
                       alpha=0.25)
# plt.figure(figsize=(100, 60), dpi=80, facecolor='w', edgecolor='k')
plt.show()
```
