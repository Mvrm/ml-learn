### Importing dataset from Sklearn

```
from sklearn.datasets import load_iris
import pandas as pd
```

```
dataset = load_iris()
```

```
df = pd.DataFrame(data=dataset.data,    # values
            columns=dataset.feature_names)  # 1st row as the column names

df['target'] = dataset.target
```

```
df.target.value_counts().head()
```

### Styling

```
def color_map(val):
  color = 'red' if val<0.5 else 'black'
  return 'color: %s' %color

df.head().style.applymap(color_map)
```
