```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import roc_auc_score
from sklearn.calibration import  CalibratedClassifierCV, calibration_curve
```

#### Toy dataset

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Store the feature data
X = data.data
# store the target data
y = data.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
clf = SGDClassifier(penalty='l1', alpha=0.001, tol=0.01, max_iter=5000)
```

```python
clf.fit(X_train, y_train)
```

#### Diagnose calibration

```python
prob= clf.decision_function(X_test)
```

```python
fop, apv= calibration_curve(y_test, prob, n_bins=10, normalize=True) 
```

```python
plt.plot([0,1], [0,1])
plt.plot(apv,fop)
plt.grid()
plt.xlabel("Avg. Probability")
plt.ylabel("Fraction of Positive")
plt.title("Calibration Plot")
plt.show()
```

#### sigmoid calibration 
- not working

```python
#using calibrated classifier cv
clf = SGDClassifier(penalty='l1', alpha=0.001, tol=0.01, max_iter=5000)
calibrated_model= CalibratedClassifierCV(clf, method='sigmoid', cv=5)
calibrated_model.fit(X_train, y_train)
```

```python
prob= calibrated_model.predict_proba(X_test)[:,1]
```

```python
fop, apv= calibration_curve(y_test, prob, n_bins=10, normalize=True)
```

```python
plt.plot([0,1], [0,1])
plt.plot(apv,fop)
plt.grid()
plt.xlabel("Avg. Probability")
plt.ylabel("Fraction of Positive")
plt.title("Calibration Plot")
plt.show()
```
